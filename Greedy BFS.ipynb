```python
import heapq
def get_graph():
    graph = {}
    num_edges = int(input("Enter the number "))
    print("Enter the edges in format node1 node2 cost:")
    for _ in range(num_edges):
        u,v,cost = input().split()
        cost = int(cost)
        if u in graph:
            graph[u].append((v,cost))
            
        else:
            graph[u] = [(v,cost)]
            
        if v not in graph:
            graph[v] = []
    return graph


def get_hueristic():
    hueristic = {}
    n = int(input("Enter the number of vertices"))
    print("Enter the hueristics in  the format node hueristic:")
    for _ in range(n):
        node,h = input().split()
        h = int(h)
        hueristic[node] = h
    return hueristic


def greedy_bfs(graph,hueristic,start,goal):
    pq = []
    heapq.heappush(pq,(hueristic[start],start))
    visited = set()
    parent = { start : None}
    
    while pq:
        _,current = heapq.heappop(pq)
        
        if current == goal:
            path = []
            while current:
                path.append(current)
                current = parent[current]
            path.reverse()
            return path
        visited.add(current)
        for neighbour,cost in graph[current]:
            if neighbour not in visited:
                heapq.heappush(pq,(hueristic[neighbour],neighbour))
                parent[neighbour] = current
                
    return None


def main():
    graph = get_graph()
    hueristics = get_hueristic()
    start = input('Enter the start vertex: ')
    goal = input('Enter the goal: ')
    
    path = greedy_bfs(graph,hueristics,start,goal)
    if path:
        print("Path found: "," -> ".join(path))
    else:
        print("Path not found")
        
        
if __name__  == "__main__":
    main()
            
    
```
